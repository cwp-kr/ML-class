```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data0405.csv', usecols=['x']).values
y = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data0405.csv', usecols=['y']).values

plt.figure(figsize = (6, 4))
plt.plot(x, y, 'bo')
plt.title('Training Data', fontsize = 16)
plt.xlabel('X', fontsize = 12)
plt.ylabel('Y', fontsize = 12)
plt.grid(alpha = 0.4)
plt.show()
```

```python
A = np.hstack([x**0, x])
A = np.asmatrix(A)

w = (A.T*A).I*A.T*y
w
```

```python
x = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data0405.csv', usecols=['x']).values
y = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data0405.csv', usecols=['y']).values

plt.figure(figsize = (6, 4))
plt.plot(x, y, 'bo')
plt.xlabel('X', fontsize = 12)
plt.ylabel('Y', fontsize = 12)

xp = np.arange(-2, 1, 0.01).reshape(-1,1)
yp = w[0,0] + w[1,0]*xp

plt.title('linear Regression', fontsize = 16)
plt.plot(xp, yp, 'r', linewidth = 2, label = "Linear model")
plt.grid(alpha = 0.4)
plt.show()
```

```python
A1 = np.hstack([x**0, x, x**2])
A1 = np.asmatrix(A1)

w1 = (A1.T*A1).I*A1.T*y
w1
```

```python
x = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data0405.csv', usecols=['x']).values
y = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data0405.csv', usecols=['y']).values

plt.figure(figsize = (6, 4))
plt.plot(x, y, 'bo')
plt.xlabel('X', fontsize = 12)
plt.ylabel('Y', fontsize = 12)

xp = np.arange(-2, 1, 0.01).reshape(-1, 1)
yp1 = w1[0,0] + w1[1,0]*xp + w1[2,0]*xp**2

plt.title('Nonlinear Regression 2', fontsize = 16)
plt.plot(xp, yp1, 'r', linewidth = 2, label = "Linear model")
plt.grid(alpha = 0.4)
plt.show()
```

```python
d = 11

xp = np.arange(-2, 1, 0.01).reshape(-1, 1)
polybasis = np.hstack([xp**i for i in range(d)])
polybasis = np.asmatrix(polybasis)

A2 = np.hstack([x**i for i in range(d)])
A2 = np.asmatrix(A2)
w11 = (A2.T*A2).I*A2.T*y
w11
```

```python
x = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data0405.csv', usecols=['x']).values
y = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data0405.csv', usecols=['y']).values

plt.figure(figsize = (6, 4))
plt.plot(x, y, 'bo')
plt.xlabel('X', fontsize = 12)
plt.ylabel('Y', fontsize = 12)

xp = np.arange(-2, 1, 0.01).reshape(-1, 1)
yp11 = polybasis*w11

plt.title('Nonlinear Regression 10', fontsize = 16)
plt.plot(xp, yp11, 'r', linewidth = 2, label = "Linear model")
plt.grid(alpha = 0.4)
plt.show()
```

```python
w2 = 0
w3 = 0

n = float(len(x))

for i in range(500):
    y_pred = w3*x + w2
    dw3 = (1/n) * sum(x * (y_pred-y))
    dw2 = (1/n) * sum(y_pred-y)
    w3 = w3 - 0.001 * dw3
    w2 = w2 - 0.001 * dw2
print (w3, w2)
```

```python
x = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data0405.csv', usecols=['x']).values
y = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data0405.csv', usecols=['y']).values

plt.figure(figsize = (6, 4))
plt.plot(x, y, 'bo')
plt.xlabel('X', fontsize = 12)
plt.ylabel('Y', fontsize = 12)

xp = np.arange(-2, 1, 0.01).reshape(-1, 1)
y_pred = w3*xp + w2

plt.title('Linear Regression(Gradient Descent)', fontsize = 16)
plt.plot(xp, y_pred, 'r', linewidth = 2, label = "Linear model")
plt.grid(alpha = 0.4)
plt.show()
```

```python
m0 = 0
m1 = 0
m2 = 0

n = float(len(x))

for i in range(500):
    y_pred1 = m2*x**2 + m1*x + m0
    dm2 = (1/n) * sum(x**2 * (y_pred1-y))
    dm1 = (1/n) * sum(x * (y_pred1-y))
    dm0 = (1/n) * sum(y_pred1-y)
    m2 = m2 - 0.001 * dm2
    m1 = m1 - 0.001 * dm1
    m0 = m0 - 0.001 * dm0
print (m2, m1, m0)
```

```python
x = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data0405.csv', usecols=['x']).values
y = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data0405.csv', usecols=['y']).values

plt.figure(figsize = (6, 4))
plt.plot(x, y, 'bo')
plt.xlabel('X', fontsize = 12)
plt.ylabel('Y', fontsize = 12)

xp = np.arange(-2, 1, 0.01).reshape(-1, 1)
y_pred1 = m2*xp**2 + m1*xp + m0

plt.title('Nonlinear Regression(Gradient Descent 2)', fontsize = 16)
plt.plot(xp, y_pred1, 'r', linewidth = 2, label = "Linear model")
plt.grid(alpha = 0.4)
plt.show()
```

```python
j0 = 0
j1 = 0
j2 = 0
j3 = 0
j4 = 0

n = float(len(x))

for i in range(500):
    y_pred1 = j4*x**4 + j3*x**3 + j2*x**2 + j1*x + j0
    dj4 = (1/n) * sum(x**4 * (y_pred1-y))
    dj3 = (1/n) * sum(x**3 * (y_pred1-y))
    dj2 = (1/n) * sum(x**2 * (y_pred1-y))
    dj1 = (1/n) * sum(x * (y_pred1-y))
    dj0 = (1/n) * sum(y_pred1-y)
    j4 = j4 - 0.001 * dj4
    j3 = j3 - 0.001 * dj3
    j2 = j2 - 0.001 * dj2
    j1 = j1 - 0.001 * dj1
    j0 = j0 - 0.001 * dj0
print (j4, j3, j2, j1, j0)
```
