```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
```

```python
X = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/boston_house_data.csv', usecols=['12']).values
y = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/boston_house_target.csv', usecols=['0']).values
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.3)
```

```python
plt.figure(figsize = (6, 6))
plt.title('linear Regression', fontsize = 16)
plt.plot(X_train, y_train ,'bo', markersize=4, label='Train data')
plt.plot(X_test, y_test,'go', markersize=4, label='Test data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
w2 = 0
w3 = 0

n = float(len(X_train))

for i in range(10000):
    y_pred = w3*X_train + w2
    dw3 = (1/n) * sum(X_train * (y_pred-y_train))
    dw2 = (1/n) * sum(y_pred-y_train)
    w3 = w3 - 0.001 * dw3
    w2 = w2 - 0.001 * dw2
print (w3, w2)
```

```python
xp = np.arange(0, 35, 0.01).reshape(-1, 1)
y_pred = w3*xp + w2

plt.figure(figsize = (6, 6))
plt.title('linear Regression', fontsize = 16)
plt.plot(X_train, y_train ,'bo', markersize=4, label='Train data')
plt.plot(X_test, y_test,'go', markersize=4, label='Test data')
plt.plot(xp, y_pred, 'r', linewidth = 2)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
clf = linear_model.LinearRegression()
clf.fit(X_train, y_train)
```

비선형

```python
m0 = 0
m1 = 0
m2 = 0

n = float(len(X_train))

for i in range(1000):
    y_pred1 = m2*X_train**2 + m1*X_train + m0
    dm2 = (1/n) * sum(X_train**2 * (y_pred1-y_train))
    dm1 = (1/n) * sum(X_train * (y_pred1-y_train))
    dm0 = (1/n) * sum(y_pred1-y_train)
    m2 = m2 - 0.00001 * dm2
    m1 = m1 - 0.00001 * dm1
    m0 = m0 - 0.00001 * dm0
print (m2, m1, m0)
```

```python
xp = np.arange(0, 35, 0.01).reshape(-1, 1)
y_pred1 = m2*xp**2 + m1*xp + m0

plt.figure(figsize = (6, 6))
plt.title('linear Regression', fontsize = 16)
plt.plot(X_train, y_train ,'bo', markersize=4, label='Train data')
plt.plot(X_test, y_test,'go', markersize=4, label='Test data')
plt.plot(xp, y_pred1, 'r', linewidth = 2)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


regressor = LinearRegression()
regressor.fit(X_train, y_train)


y_pred = regressor.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)


print("Predictions:", y_pred)
print("Mean Squared Error (MSE):", mse)
print("R² Score:", r2)
```

```python
def h(x,w):
    return 1/(1 + np.exp(-x@w))
```

```python
w = np.zeros([3,1])

alpha = 0.01

for i in range(10000):
    df = -X.T@(y - h(X,w))
    w = w - alpha*df

print(w)
```
